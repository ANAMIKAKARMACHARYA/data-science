# Line chart

```python
import pandas as pd
from matplotlib import pyplot as plt
years = [2000,2004,2008,2012,2016]
gdp=[1212,2323,1414,4455,5600]
# create a line chart ,years on x-axis ,gda on y-axis
plt.plot(years, gdp,color='orange',marker='*',linestyle='dotted')
# add a title
plt.title("Nominal GDP")
# add a label to the y axis
plt.ylabel('Billions of $')
plt.show()
```

##### The line chart represents nominal GDP growth over the years (2000-2016).The x-axis represents years and the y axis represents GDP in billions of dollars.The GDP generally increases over time,showing economic growth,except for the year 2008.From the year 2000 to 2004,GDP increases from 1212 billion dollars to 2323 billion dollars.In 2008,GDP drastically decreased to 1414 billion dollars reflecting a sharp economic downturn which may be due to financial crisis.But the GDP recovered significantly to 4455 billion dollars by 2012 and continued to grow to 5600 billion dollars,showing the strong economic recovery.

# Pie chart

```python
import pandas as pd
import seaborn as sns
df=pd.read_csv("./Health_heart_experimental.csv")
df.head(11)
df['sex'].value_counts().plot(kind='pie',autopct='%0.2f')
```

##### The pie chart above shows the count of male and female population of the health heart experiment.By observing the piechart it reflects that there is slightly more count of male population i.e 50.09% than of female population which is 49.91%. This indicates that both male and female participants were almost equally represented in the experiment, with a very small difference in their participation.

# Box plot

```python
sns.boxplot(df['age'])
```

##### The above box plot visually illustrates the distribution of the age column within the health heart experimental dataset.It reveals that the minimum age is 20 where as the maximum age is 69.The 25th percentile (Q1) indicates that 25% of participants are around 32 years old. The median (50th percentile) age, which represents the middle of the population, falls between 40 to 45 years, meaning half of the participants are in this age group. Furthermore, the 75th percentile (Q3) shows that 75% of the participants are aged between 55 to 60 years and there are no outliers present which means there is no data points which fall outside this range.  

# Histogram

```python
plt.hist(df['BMI'])
```

##### The above histogram represents the distribution of Body Mass Index values of the dataset.BMI values range from approximately 13 to 40,where lower values suggest underweight individuals and higher value correspond to overweight.The tallest bars, peaking between a BMI of about 26.5 to 29.2, indicate that most individuals in the dataset fall in this range, categorizing them as overweight. The histogram also shows fewer individuals at the extremes of the BMI scale (underweight and overweight), suggesting the data has a higher concentration of values near the middle range.

# Scatter plot

```python
tips = sns.load_dataset('tips')
print(tips)
```

```python
sns.scatterplot(x='total_bill', y='tip',hue='sex',style='smoker', data=tips)
plt.show()
```

##### The scatter plot illustrates the relationship between the total bill and tip of a restraurant.The plot shows that the highest tip,which is 10 dollar is given by a male customer at a total bill of 50 dollar.The lowest tip, ranging from 0 to 2 dollar is given by a female customer with a bill of less than 10 dollar.A noticable trend is that many customer tends to leave tips in the range of 2 to 4 dollar for bills between 10 to 20 dollar.Beside that the plot suggest that the higher propotion of female customers appear to be smokers.
